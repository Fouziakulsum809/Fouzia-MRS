```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

data cleaning

```python
d.drop(['has_health_insur'],axis=1,inplace=True)
```

```python
d.drop(['housing_status'],axis=1,inplace=True)
```

```python
d['h1n1_worry']=d['h1n1_worry'].replace(np.nan,d['h1n1_worry'].mode()[0])
```

```python
d['h1n1_awareness']=d['h1n1_awareness'].replace(np.nan,d['h1n1_awareness'].mode()[0])
```

```python
d['antiviral_medication']=d['antiviral_medication'].replace(np.nan,d['antiviral_medication'].mode()[0])
```

```python
d['contact_avoidance']=d['contact_avoidance'].replace(np.nan,d['contact_avoidance'].mode()[0])
```

```python
d['bought_face_mask']=d['bought_face_mask'].replace(np.nan,d['bought_face_mask'].mode()[0])
```

```python
d['wash_hands_frequently']=d['wash_hands_frequently'].replace(np.nan,d['wash_hands_frequently'].mode()[0])
```

```python
d.dropna(inplace=True)
```

```python
d.isnull().sum()
```

```python
d.head()
```

```python
d.drop(['unique_id'],axis=1,inplace=True)
```

ml model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x=d.drop(['h1n1_vaccine'],axis=1)
y=d['h1n1_vaccine']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
l=LogisticRegression()
l.fit(x_train,y_train)
```

```python
l.score(x_train,y_train)
```

```python
l.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
de=DecisionTreeClassifier(max_depth=5)
de.fit(x_train,y_train)
```

```python
de.score(x_train,y_train)
```

```python
de.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=3)
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
rfc.score(x_train,y_train)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
```

```python
svc.score(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
knn.score(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
